```python
import pandas as pd
import numpy as np

Password = pd.read_csv("dfWordCount")
```

```python
Password["FChar"] = Password.password.astype(str).str[0]
Password["LChar"] =Password.password.astype(str).str[-1]
```

```python
def n_lower_chars(string):
    return sum(map(str.islower, string))
def n_upper_chars(string):
    return sum(map(str.isupper, string))
def n_num_chars(string):
    return sum(map(str.isdigit, string))
def n_space_chars(string):
    return sum(map(str.isspace, string))
```

```python
Password['TotalLength'] = Password.password.astype(str).apply(len)
Password['LowerChar'] = Password.password.astype(str).apply(n_lower_chars)
Password['UpperChar'] = Password.password.astype(str).apply(n_upper_chars)
Password['NumChar'] = Password.password.astype(str).apply(n_num_chars)
Password['SpaceChar'] = Password.password.astype(str).apply(n_space_chars)
```

```python
Password.groupby('SpaceChar').count()
```

```python
Password['SpecialChar'] = Password.TotalLength - (Password.LowerChar + Password.UpperChar + Password.SpaceChar + Password.NumChar)
```

```python
data = Password
strlen = data.password.str.len()
data["Length"] = strlen
```

```python
from collections import Counter
import re
temp = re.findall(r'\d+', data.password[0]) 
res = bool(map(int, temp)) 
print(res)
```

```python
def split_string(string):
    return list(str(string))

def reps(x):
    temp = split_string(str(x))
    i = 0
    j = 1
    k = 2
    flag = 0
    for y in range(2,len(temp)-1):
        if temp[i] == temp[j] and temp[j] == temp[k]:
            flag = 1
            x = len(temp) + 1
        else:
            i += 1
            j += 1
            k += 1
    return flag
```

```python
data['Repeats'] = data.password.astype(str).apply(reps)
```
